```python
ogrenciler = ("Ahmet", "Mehmet", "Ayşe", "Zeynep")

notlar = {
"Ahmet": [78,85,90],
"Mehmet": [60,72,70],
"Ayşe": [90,95,88],
"Zeynep": [50,65,60]
}

ortalama = {}
ortalama["Ahmet"] = sum(notlar["Ahmet"]) / len(notlar["Ahmet"])
ortalama ["Mehmet"] = sum(notlar["Mehmet"]) / len(notlar["Mehmet"])                                              
ortalama["Ayşe"] = sum(notlar["Ayşe"]) / len(notlar["Ayşe"]) 
ortalama["Zeynep"] = sum(notlar["Zeynep"]) / len(notlar["Zeynep"])

durum = {}
durum["Ahmet"] = ortalama["Ahmet" ]>= 70
durum["Mehmet"] = ortalama["Mehmet"] >= 70
durum["Ayşe"] = ortalama["Ayşe"] >= 70
durum["Zeynep"] = ortalama["Zeynep"] >= 70

en_iyi_ort = max(ortalama.values())
en_kotu_ort = min(ortalama.values())

#isimleri karşılaştırıyoruz ama ortalaması yüksek/düşük olanı getlemek 
en_iyi_ogr_isim = max(ortalama, key = ortalama.get)
en_iyi_ogr = (en_iyi_ogr_isim, round(en_iyi_ort, 1))

en_kotu_ogr_isim = min(ortalama, key = ortalama.get)
en_kotu_ogr = (en_kotu_ogr_isim, round(en_kotu_ort,1))

#ilk iki ders alma
ahmet_ilkikiders = notlar["Ahmet"][:2]
mehmet_ilkikiders = notlar["Mehmet"][:2]
ayse_ilkikiders = notlar["Ayşe"][:2]
zeynep_ilkikiders = notlar["Zeynep"][:2]

#tüm notlar tek listede
tüm_notlar = notlar["Ahmet"] + notlar["Mehmet"] + notlar["Ayşe"] + notlar["Zeynep"]
set_notlar = set(tüm_notlar)

#ilk üç harf
ahm_ucharf = ogrenciler[0].upper()[:3]
meh_ucharf = ogrenciler[1].upper()[:3]
ays_ucharf = ogrenciler[2].upper()[:3]
zey_ucharf = ogrenciler[3].upper()[:3]

#sınıf_ort = (ortalama["Ahmet"] + ortalama["Mehmet"] + ortalama["Ayşe"] + ortalama["Zeynep"]) / len(ogrenciler)
sınıf_ort = sum(ortalama.values()) / len(ortalama)

basarılı_ogrler = []
if durum ["Ahmet"]: basarılı_ogrler.append("Ahmet")
if durum ["Mehmet"] : basarılı_ogrler.append("Mehmet")
if durum ["Ayşe"] : basarılı_ogrler.append("Ayşe")
if durum ["Zeynep"] : basarılı_ogrler.append("Zeynep")    

print ("ÖĞRENCİ ANALİZ RAPORU\n") 
print (f"{ahm_ucharf} -> Ortalaması : {ortalama["Ahmet"] :.1f} = {"Geçti" if durum["Ahmet"] else "Kaldı"}")
print(f"{meh_ucharf} -> Ortalaması : {ortalama["Mehmet"] :.1f} = {"Geçti" if durum["Mehmet"] else "Kaldı"}")       
print(f"{ays_ucharf} -> Ortalaması : {ortalama["Ayşe"] :.1f} = {"Geçti" if durum["Ayşe"] else "Kaldı"}")
print(f"{zey_ucharf} -> Ortalaması : {ortalama["Zeynep"] :.1f} = {"Geçti" if durum["Zeynep"] else "Kaldı\n"}")

print(f"En başarılı öğrenci : {en_iyi_ogr}")
print(f"En başarısız öğrenci : {en_kotu_ogr}")

print(f"\nSınıf ortalaması : {sınıf_ort :.1f}")
print(f"Tekil not sayısı : {len(set_notlar)}")
```
